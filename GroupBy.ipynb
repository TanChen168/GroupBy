```python
import pandas as pd
filename = '/content/sample_data/CalIt2.csv'

df = pd.read_csv(filename)

df.info()
```

RENAME COLUMNS


```python
df.columns=['Flow_ID', 'Date', 'Time', 'Count']
df
```

SELECTING DATA - Select all data for the date July 24 2005 having flow id=7

```python
date_filter = df['Date'] == '07/24/05'
flow_filter = df['Flow_ID'] == 7
df = df.loc[date_filter & flow_filter, :]
df
```

SELECTING DATA - From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10.

```python
var_top10 = df.sort_values('Count', axis=0, ascending=False).head(10)
var_top10
```

APPLY FUNCTION - For the 10 rows outputted above, use Pandas Apply function to subtract lowest value of the 10 from all of them and then output the average value of the resulting counts

```python
min_count = var_top10['Count'].min()
min_count
```

```python
var_top10['Count'].mean()
var_top10
```

```python
def fn_new_count(a):
  return a - min_count

x = var_top10.apply(lambda  row: fn_new_count(var_top10['Count']))
x['Count'].mean()
```

GROUPING - Select data in the month of August 2005 having flow id=7

```python
from datetime import datetime
df['Date'] = pd.to_datetime(df['Date'])
month_filter = df['Date'].dt.month == 8
year_filter = df['Date'].dt.year == 2005
flow_filter = df['Flow_ID'] == 7
df_months = df.loc[month_filter & year_filter, :]
df_months.sort_values(by=['Count'])
```

GROUPING - Group the data based on date and get the max count per date

```python
df.groupby(['Date']).max()
```
